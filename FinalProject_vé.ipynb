```python
attend = pd.read_csv("NFL_Attendance_2000_2019.csv")
```

```python
attend
```

```python
# Dictionary to map team names to abbreviations
team_abbr = {
    'Cardinals': 'ARI',
    'Falcons': 'ATL',
    'Ravens': 'BAL',
    'Bills': 'BUF',
    'Panthers': 'CAR',
    'Bears': 'CHI',
    'Bengals': 'CIN',
    'Browns': 'CLE',
    'Cowboys': 'DAL',
    'Broncos': 'DEN',
    'Lions': 'DET',
    'Packers': 'GB',
    'Texans': 'HOU',
    'Colts': 'IND',
    'Jaguars': 'JAX',
    'Chiefs': 'KC',
    'Raiders': 'LV',
    'Chargers': 'LAC',
    'Rams': 'LAR',
    'Dolphins': 'MIA',
    'Vikings': 'MIN',
    'Patriots': 'NE',
    'Saints': 'NO',
    'Giants': 'NYG',
    'Jets': 'NYJ',
    'Eagles': 'PHI',
    'Steelers': 'PIT',
    '49ers': 'SF',
    'Seahawks': 'SEA',
    'Buccaneers': 'TB',
    'Titans': 'TEN',
    'Redskins': 'WAS'
}

# Extract the last name from team names
attend['last_name'] = attend['team_name'].apply(lambda x: x.split()[-1])

# Map team names to abbreviations
team_abbr_map = {team: team_abbr[team] for team in attend['last_name'].unique()}

# Create a new column with abbreviations
attend['team_abbr'] = attend['last_name'].map(team_abbr_map)

# Drop the 'last_name' column if not needed
# df.drop(columns=['last_name'], inplace=True)

# Print the DataFrame to verify the new column
print(attend)
```

```python

# Create a Folium map centered on the USA
map = folium.Map(location=[39.8283, -98.5795], zoom_start=4)

for index, stadium in stadium_coord.iterrows():
    # Access latitude, longitude, and name from each row
    lat = stadium["Latitude"]
    lon = stadium["Longitude"]
    name = stadium["StadiumName"]
    
    # Add marker to the map
    folium.Marker(location=[lat, lon], tooltip=name).add_to(map)

# Display the map
map
```
